# Marklund 1988--calculation of biomass

```python
import numpy as np
import pandas as pd

def biomass_M1988(dbh_cm, height_m, spp, this_period):
    # Convert inputs to cm and meters, respectively
    d = dbh_cm
    H = height_m 
    
    # Initialize a dictionary to hold biomass components
    biomass_kg = {'crown': None, 'dgr': None, 'stem': None, 'bark': None,
                  'stub': None, 'usoil': None, 'rot1': None, 'rot2': None,
                  'foliage': None, 'stem2': None, 'living_branches': None, 'dead_branches': None}
    
    # Check species type
    if spp == "spruce":
        # Calculate biomass components using Marklund's equations (1988)
        dgr = np.exp(-4.6351 + 3.6518 * (d / (d + 18)) + 0.0493 * H + 1.0129 * (np.log(H)))  #G20
        crown = np.exp(-1.2063 + 10.9708 * (d / (d + 13)) - 0.0124 * H - 0.4923 * (np.log(H))) #G12   #living branches including needles
        foliage = np.exp(-1.8551 + 9.7809 * (d / (d + 12)) - 0.4873 * (np.log(H)))          #G16 #needles
        bark = np.exp(-3.4020 + 8.3089 * (d / (d + 15)) + 0.0147 * H + 0.2295 * (np.log(H)))   #G8
        stem = np.exp(-2.3032 + 7.2309 * (d / (d + 14)) + 0.0355 * H + 0.7030 * (np.log(H))) #G5
        dead_branches = np.exp(-4.6351 + 3.6518 * (d / (d + 18))+ 0.0493*H + 1.0129*(np.log(H))) #G20
        stub = np.exp(-3.3645 + 10.6686 * (d / (d + 17)))  #G26
        rot1 = np.exp(-2.5706 + 7.6283 * (d / (d + 12)))   #G31
        rot2 = np.exp(-6.3851 + 13.3703 * (d / (d + 8)))    #G28


        ## Ina
        stem2 = np.exp(-2.0571 + 11.3341 * (d / (d + 14))) #G1
        living_branches = np.exp(-1.2804 + 8.5242 * (d / (d + 13))) #G11
        #dead_branches = np.exp(-4.3308 + 9.955 * (d / (d + 18))) #G19
        
        # Petterson and Sthål (2006) equation for roots < 2mm
        usoil = (np.exp(((0.32308)**2) / 2 + 4.58761 + d * 10 /
                 (d * 10 + 138) * 10.44035)) / 1000
        
        # Update the biomass_kg dictionary
        biomass_kg.update({'crown': crown, 'dgr': dgr, 'stem': stem, 'bark': bark,
                           'stub': stub, 'usoil': usoil, 'rot1': rot1, 'rot2': rot2,
                           'foliage': foliage, 'stem2': stem2, 'living_branches': living_branches, 'dead_branches': dead_branches})
    
    # Correct biomass components to zero for d and H equal to zero
    for key, value in biomass_kg.items():
        if d == 0 or H == 0:
            biomass_kg[key] = 0
    
    return biomass_kg

# Example usage
# read data from a csv file
Ulf = pd.read_excel(r'./zhicong/Ulf_biomass_origin.xlsx')
# extract the dbh and height
dbh_cm = Ulf['t_dbh_after_thinning']
height_m = Ulf['t_height_after_thinning']
tree_no = Ulf['t_n_after_thinning']
spp_example = "spruce"
this_period_example = "t1"  # This variable is not used in this snippet but included for completeness

for i in range(len(tree_no)):
    #print(f"Tree {tree_no[i]}")
    biomass_result = biomass_M1988(dbh_cm[i], height_m[i], spp_example, this_period_example)
    # sum up the above ground biomass components
    abg_components = [biomass_result[key] for key in biomass_result.keys() if key == 'crown' or key == 'bark' or key == 'stem' or key == 'dead_branches']
    #Ina
    #abg_components = [biomass_result[key] for key in biomass_result.keys() if key == 'stem2' or key == 'living_branches' or key == 'dead_branches']
    bgb_components = [biomass_result[key] for key in biomass_result.keys() if key == 'stub' or key == 'rot1' or key == 'rot2' ]#key == 'usoil' or
    # save the result to a new already created column "AGB_Marklund" and BGB_Marklund, and convrt to Mg/ha, and save it in the excel file
    Ulf.loc[i, 'AGB_Marklund'] = sum(abg_components*tree_no[i])/1000
    Ulf.loc[i, 'BGB_Marklund'] = sum(bgb_components*tree_no[i])/1000
# save the result to an existing excel file
Ulf.to_excel(r'./zhicong/Ulf_biomass_origin.xlsx', index=False)  # Save the data to a new
# print the ABG and BGB columns (in Mg/ha), top 10 rows

# biomass_result = biomass_M1988(dbh_cm, height_m, spp_example, this_period_example)
# print(biomass_result)
```

```python
import pandas as pd
import numpy as np

# Load the data from the provided path
file_path = './zhicong/Ulf_biomass_origin.xlsx'
ulf = pd.read_excel(file_path)

# Convert 'DATE' to datetime if necessary
ulf['DATE'] = pd.to_datetime(ulf['DATE'])
ulf['Year'] = ulf['DATE'].dt.year

# Define the treatment groups
treatment_groups = {
    'Control': [3, 15, 18, 24],
    'Irrigation': [4, 12, 14, 25],
    'D2': [42, 44, 46, 48],
    'D1': [41, 43, 45, 47]
}

# Initialize an empty DataFrame to hold the aggregated data
aggregated_data = pd.DataFrame()

# Process each treatment group
for treatment_name, plot_numbers in treatment_groups.items():
    # Filter data for the current treatment
    treatment_data = ulf[ulf['plot no'].isin(plot_numbers)]
    
    # Group by year and calculate mean 'AGB_Marklund' for each group
    yearly_means = treatment_data.groupby('Year')['AGB_Marklund'].mean().reset_index()
    
    # If the aggregated_data is empty, initialize it with the year and current treatment's data
    if aggregated_data.empty:
        aggregated_data = yearly_means.rename(columns={'AGB_Marklund': treatment_name})
    else:
        # Merge the current treatment's data with the aggregated data
        aggregated_data = pd.merge(aggregated_data, yearly_means, on='Year', how='outer').rename(columns={'AGB_Marklund': treatment_name})

# Ensure all NaN values are filled with 0 or another appropriate placeholder
aggregated_data.fillna(value=np.nan, inplace=True)
# Save the structured_data to a new CSV file
output_file_path = './structured_biomass_data.csv'
aggregated_data.to_csv(output_file_path, index=False)

print(f"Data saved to {output_file_path}")
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Data for C Stem
years = np.array([1987, 1988, 1989, 1990, 1991, 1992, 1993])
c_stem = np.array([3035.676132, 3306.403064, 3512.09532, 3758.843445, 3902.447178, 4101.608247, 4275.232437])

# Data for C Leaf
c_leaf = np.array([586.0075092, 629.8946633, 663.0508962, 702.1970221, 724.7326525, 755.9858789, 782.9237071])

# Define the exponential model
def exp_model(x, a, b):
    return a * np.exp(b * (x - min(years)))

# Fit the model for C Stem
params_stem, _ = curve_fit(exp_model, years, c_stem, p0=[1, 0.01])
a_stem, b_stem = params_stem

# Fit the model for C Leaf
params_leaf, _ = curve_fit(exp_model, years, c_leaf, p0=[1, 0.01])
a_leaf, b_leaf = params_leaf

# Predict C Stem and C Leaf for 1966
year_1966 = 1966
c_stem_1966 = exp_model(year_1966, a_stem, b_stem)
c_leaf_1966 = exp_model(year_1966, a_leaf, b_leaf)

# Display the results
c_stem_1966, c_leaf_1966

# Plotting the data and the fitted model
plt.figure(figsize=(12, 6))

# Plot for C Stem
plt.subplot(1, 2, 1)
plt.scatter(years, c_stem, color='blue', label='Data')
plt.plot(years, exp_model(years, a_stem, b_stem), color='red', label=f'Exp Model: a * exp(b * (x - {min(years)}))')
plt.xlabel('Year')
plt.ylabel('C Stem')
plt.title('C Stem vs Year')
plt.legend()

# Plot for C Leaf
plt.subplot(1, 2, 2)
plt.scatter(years, c_leaf, color='green', label='Data')
plt.plot(years, exp_model(years, a_leaf, b_leaf), color='red', label=f'Exp Model: a * exp(b * (x - {min(years)}))')
plt.xlabel('Year')
plt.ylabel('C Leaf')
plt.title('C Leaf vs Year')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import pandas as pd

# Assuming 'all_variables' is already populated with the data as per previous steps

def calculate_average(data):
    # Convert 'Date' to datetime format for proper plotting and grouping
    data['Date'] = pd.to_datetime(data['Date'])
    
    # Ensure 'tensiometrat' is numeric, converting non-numeric to NaN
    data['tensiometrat'] = pd.to_numeric(data['tensiometrat'], errors='coerce')

    # Group by date and calculate the average 'tensiometrat'
    avg_data = data.groupby('Date')['tensiometrat'].mean().reset_index()
    
    return avg_data

# Dictionary to store the average data frames
average_data = {}

# Treatment groups and their corresponding plots
treatments = {
    'I': ['12I', '25I', '24I'],
    'D1': ['8D1', '16D1', '23D1'],
    'D2': ['8D2', '16D2', '23D2'],
    'C': ['24C', '3C', '15C']
}

# Depth categories
depths = [15, 30, 45, 60]

# Process each treatment and depth
for treatment, plots in treatments.items():
    for depth in depths:
        variable_name = f"{treatment}-{depth}"
        data_subset = all_variables.get(variable_name, pd.DataFrame())  # Get the data or an empty DataFrame if not present
        
        # Calculate average if there is data available
        if not data_subset.empty:
            avg_df = calculate_average(data_subset)
            
            # Store in dictionary with a new variable name indicating averaging
            avg_var_name = f"{treatment}-{depth}-avg"
            average_data[avg_var_name] = avg_df

# Example: Accessing averaged data for D1 at 15 cm depth
print(average_data.get('D1-15-avg', 'No data available'))

```

```python
# make Forcing data for D1 and D2 files
import pandas as pd

# Load the CSV file
file_path = r'.//zhicong/SMHI/Skogaby_weather_Zhicong.csv'  # Replace with the actual file path
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format for easy filtering
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M')

# Create two copies of the dataframe for the two new files
df_1 = df.copy()
df_2 = df.copy()

# Define the ranges and apply the transformation
# For df_1 (April-September 1988-1989)
mask_1 = (df_1['Date'].dt.year.isin([1988, 1989])) & (df_1['Date'].dt.month >= 4) & (df_1['Date'].dt.month <= 9)
df_1.loc[mask_1, 'Precipitation'] = df_1.loc[mask_1, 'Precipitation'] * (1/3)

# For df_2 (April-September 1990-1996)
mask_2 = (df_2['Date'].dt.year >= 1990) & (df_2['Date'].dt.year <= 1996) & (df_2['Date'].dt.month >= 4) & (df_2['Date'].dt.month <= 9)
df_2.loc[mask_2, 'Precipitation'] = df_2.loc[mask_2, 'Precipitation'] * (1/3)

# Save the transformed dataframes to new CSV files
df_1_file_path = 'Skogaby_weather_D1_zhicong.csv'
df_2_file_path = 'Skogaby_weather_D2_zhicong.csv'

df_1.to_csv(df_1_file_path, index=False)
df_2.to_csv(df_2_file_path, index=False)

print(f"Files saved as {df_1_file_path} and {df_2_file_path}")
```

```python

#plot the data from 1990 to 2006
import matplotlib.pyplot as plt
import pandas
import matplotlib.dates as mdates

# Load the data
data = pd.read_csv('Skogaby_weather_D1_zhicong.csv')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Filter the data for the specified period
start_date = '1990-01-01'
end_date = '2006-12-31'
data_filtered = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

#calculate the cumulative precipitation year by year
data_filtered['Year'] = data_filtered['Date'].dt.year
data_filtered['Cumulative_Precipitation'] = data_filtered.groupby('Year')['Precipitation'].cumsum()

# Plotting the data
plt.figure(figsize=(14, 7))
plt.plot(data_filtered['Date'], data_filtered['Cumulative_Precipitation'], color='blue', label='Cumulative Precipitation')
plt.title('Cumulative Precipitation from 1990 to 2006')
plt.xlabel('Date')
plt.ylabel('Cumulative Precipitation (mm)')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# read control data from the excel file
import pandas as pd

# Load the data
data_control = pd.read_csv('./Zhicong/SMHI/Skogaby_weather_C_Zhicong.csv')

#calculate the cumulative precipitation year by year
data_control['Date'] = pd.to_datetime(data_control['Date'])
data_control['Year'] = data_control['Date'].dt.year
data_control['Cumulative_Precipitation'] = data_control.groupby('Year')['Precipitation'].cumsum()

# print the annual precipitation data from 1988 to 1997
annual_precipitation = data_control[data_control['Year'].between(1988, 1997)].groupby('Year')['Cumulative_Precipitation'].max()
print(annual_precipitation)

#do the same thing for temperature, but calculate the mean temperature for each year
data_control['Year'] = data_control['Date'].dt.year
annual_temperature = data_control[data_control['Year'].between(1988, 1997)].groupby('Year')['Air Temperature'].mean()
print(annual_temperature)
```
